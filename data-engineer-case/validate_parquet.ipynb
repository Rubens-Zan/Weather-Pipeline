```python


def validate_parquet_file(file_path: str):
    print("--- Running Validation Tests ---") 
    # --- Setup ---
    # Load the workload config to get the new date range (Oct 16-18)
    try:
        with open('workload.json', 'r') as f:
            workload = json.load(f)
            
        # Get the new data date range from workload.json
        BEGIN_DATE_NEW = pd.to_datetime(workload['date_config']['begin_date'], utc=True)
        END_DATE_NEW = pd.to_datetime(workload['date_config']['end_date'], utc=True)
        
        # We'll test the amsterdam file
        # file_path = "data/structured/amsterdam/202410.parquet"
        print(f"Testing file: {file_path}\n")

        df = pd.read_parquet(file_path)

        # --- Test 1: Contains both historical (Oct 1-15) and new data (Oct 16-18) ---
        print("Test 1: Checking for historical and new data...")
        min_ts = df['timestamp'].min()
        max_ts = df['timestamp'].max()
        
        # Check that data exists *before* the new data begin_date
        assert min_ts < BEGIN_DATE_NEW, f"FAIL: Historical data is missing. Min timestamp is {min_ts}"
        # Check that data exists *at or after* the new data end_date
        assert max_ts.date() >= END_DATE_NEW.date(), f"FAIL: New data is missing. Max timestamp is {max_ts}"
        print(f"  PASS: Data ranges from {min_ts} (historical) to {max_ts} (new).")
        
        # --- Test 2: No duplicate (timestamp, location) pairs ---
        print("\nTest 2: Checking for duplicates...")
        total_rows = len(df)
        unique_rows = len(df.drop_duplicates(subset=['timestamp', 'location']))
        
        assert total_rows == unique_rows, f"FAIL: Duplicates found! Total={total_rows}, Unique={unique_rows}"
        print(f"  PASS: No duplicates found ({total_rows} unique rows).")

        # --- Test 3: Timestamps in millisecond precision UTC timezone ---
        print("\nTest 3: Checking for UTC timestamps...") 
        
        assert 'utc' in str(df['timestamp'].dtype).lower(), f"FAIL: Timestamp is not UTC. Dtype is {df['timestamp'].dtype}"
        print(f"  PASS: Timestamps are UTC ({df['timestamp'].dtype}).")
        
        # --- Test 4: All columns from both historical and new data present ---
        print("\nTest 4: Checking for merged schema...")
        columns = df.columns
        # 'dew_point' is from the historical file
        assert 'dew_point' in columns, "FAIL: 'dew_point' column from historical data is missing."
        # 'temperature_2m' is from the new data (via workload.json)
        assert 'temperature_2m' in columns, "FAIL: 'temperature_2m' column from new data is missing."
        print(f"  PASS: Schema is merged. Found 'dew_point' and 'temperature_2m'.")
        
        print("\n--- All Validation Checks Passed! ---")
        
    except FileNotFoundError:
        print(f"ERROR: Test file not found at {file_path}. Did the 'transform' stage run?")
    except AssertionError as e:
        print(f"\n--- TEST FAILED ---")
        print(f"  {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```

```python
validate_parquet_file('data/structured/amsterdam/202410.parquet')
validate_parquet_file('data/structured/london/202410.parquet')
```
